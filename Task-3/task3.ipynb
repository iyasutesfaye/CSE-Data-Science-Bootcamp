# PART 2 — TITANIC DATASET 

### Load data

```python
df = pd.read_csv("Titanic-Dataset.csv")
```

## Step 1: EXPLORATION

```python
df.head()
df.info()
df.describe()
```

## Head of the Dataset

The info() function provides details such as column names, data types, non-null values, and memory usage. It helps identify missing values.

##  Statistical Summary
The describe() function provides statistical information such as mean, standard deviation, minimum, maximum, and quartiles of numerical columns.

# Step 2: DATA CLEANING

## Handling Missing Values
Missing values are handled as follows:
- Age is filled with the median value
- Embarked is filled with the mode value

This ensures the dataset has no missing values in important columns.

## Removing Cabin Column
The Cabin column is removed because it contains many missing values and is not required for the analysis.

## Removing Duplicates
Duplicate rows are removed to ensure data quality and avoid redundancy in the dataset.

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# Drop Cabin safely
if "Cabin" in df.columns:
    df = df.drop(columns=["Cabin"])

# Remove duplicates
df = df.drop_duplicates()

# Check result
df.info()
df.head()
```

# 📊 Step 3: Data Analysis 
using (GroupBy)

This step focuses on analyzing relationships between variables using grouping operations to identify survival patterns.

### Survival Rate by Gender
This analysis shows how survival rates differ between male and female passengers.

```python
df.groupby("Sex")["Survived"].mean()
```

###  Survival Rate by Class
This analysis examines how passenger class (1st, 2nd, 3rd) affected survival probability.

```python
df.groupby("Pclass")["Survived"].mean()
```

###  Average Age per Class
This analysis shows the average age of passengers in each class, helping understand demographic distribution.

```python
df.groupby("Pclass")["Age"].mean()
```

####  Survival Rate by Age Group
Passengers are grouped into age categories:
- Child
- Teen
- Adult
- Senior

This helps analyze how age influences survival probability.

```python
# Create age groups
df["AgeGroup"] = pd.cut(df["Age"], bins=[0,12,18,60,100], labels=["Child","Teen","Adult","Senior"])

df.groupby("AgeGroup")["Survived"].mean()
```

# Step 4: Filtering

## Objective
Filtering is used to extract specific subsets of the dataset for more focused analysis.

###  Female Passengers Who Survived
This filter retrieves all female passengers who survived the disaster.

```python
df[(df["Sex"] == "female") & (df["Survived"] == 1)]
```

### Children Who Survived
This filter identifies passengers under the age of 12 who survived.

```python
df[(df["Age"] < 12) & (df["Survived"] == 1)]
```

###  First Class Passengers Who Survived
This filter extracts passengers from first class who survived, helping analyze high-priority groups.

```python
df[(df["Pclass"] == 1) & (df["Survived"] == 1)]
```
